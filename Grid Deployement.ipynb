```python
import matplotlib.pyplot as plt
import numpy as np

def random_deployment_with_ranges(num_sensors, area_size, max_range):
    # Randomly deploy sensors with specified communication range
    x = np.random.uniform(0, area_size, num_sensors)
    y = np.random.uniform(0, area_size, num_sensors)


    ranges = np.full(num_sensors, max_range)

    return x, y, ranges

def grid_deployment_with_ranges(num_sensors, area_size, max_range):
    side_length = int(np.sqrt(num_sensors))
    x, y = np.meshgrid(np.linspace(0, area_size, side_length), np.linspace(0, area_size, side_length))
    x = x.flatten()
    y = y.flatten()

    # Same communication range for all sensors in grid deployment
    ranges = np.full(num_sensors, max_range)

    return x, y, ranges

def plot_deployment_with_ranges(x, y, ranges, deployment_type):
    plt.scatter(x, y, label=deployment_type, c='none', edgecolors='black', marker='o')

    # Plot circles representing communication range
    for xi, yi, ri in zip(x, y, ranges):
        circle = plt.Circle((xi, yi), ri, fill=False, color='blue', linestyle='dotted')
        plt.gca().add_patch(circle)

    plt.title(f"{deployment_type} Deployment of Sensors with Ranges")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.legend()
    plt.show()

# Parameters
num_sensors = 25
area_size = 100
max_range = 20  # Maximum communication range

# Random Deployment with Ranges
random_with_ranges_x, random_with_ranges_y, random_ranges = random_deployment_with_ranges(num_sensors, area_size, max_range)
plot_deployment_with_ranges(random_with_ranges_x, random_with_ranges_y, random_ranges, "Random with Ranges")

# Grid Deployment with Ranges
grid_with_ranges_x, grid_with_ranges_y, grid_ranges = grid_deployment_with_ranges(num_sensors, area_size, max_range)
plot_deployment_with_ranges(grid_with_ranges_x, grid_with_ranges_y, grid_ranges, "Grid with Ranges")
```
