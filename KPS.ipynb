```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_grid(grid_size, sinks_positions):
    # Create a grid with sensor nodes
    grid = np.zeros((grid_size, grid_size))

    # Place sinks at specified positions
    for sink_pos in sinks_positions:
        grid[sink_pos[0], sink_pos[1]] = 2  # Use 2 to represent sinks

    # Plot sensor nodes as dots
    sensor_legend = plt.scatter([], [], color='blue', label='Sensor')  # Empty plot for creating a legend
    for i in range(grid_size):
        for j in range(grid_size):
            if grid[i, j] == 0:
                plt.scatter(j, i, color='blue')  # Use blue dots for sensor nodes

    # Plot sinks as squares
    sink_legend = plt.scatter([], [], color='red', marker='s', s=100, label='Sink')  # Empty plot for creating a legend
    for sink_pos in sinks_positions:
        plt.plot(sink_pos[1], sink_pos[0], 's', color='red', markersize=10)  # Use red squares for sinks

    # Set plot properties and add legend
    plt.title('10x10 Grid with Sensor Nodes and Sinks')
    plt.xticks(range(grid_size))
    plt.yticks(range(grid_size))
    plt.grid(True, linewidth=0.5, color='black', linestyle='--', alpha=0.5)
    plt.legend(handles=[sensor_legend, sink_legend])
    plt.show()

# Define the grid size and sink positions
grid_size = 10
sink_positions = [(0, 0), (0, grid_size-1), (grid_size-1, 0)]

# Visualize the grid with sensor nodes as dots and sinks as squares
visualize_grid(grid_size, sink_positions)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_wsn(grid_size, sinks_positions, corrupted_region, route_start, route_end, backup_route):
    # Create a grid with sensors
    grid = np.zeros((grid_size, grid_size), dtype=np.object)

    # Place sinks at specified positions
    for sink_pos in sinks_positions:
        grid[sink_pos[0], sink_pos[1]] = 'S'  # Use 'S' to represent sinks

    # Mark corrupted region
    for corr_pos in corrupted_region:
        grid[corr_pos[0], corr_pos[1]] = 'X'

    # Plot sensors as circles
    for i in range(grid_size):
        for j in range(grid_size):
            if grid[i, j] != 'S' and grid[i, j] != 'X':
                plt.scatter(j, i, color='blue', marker='o', s=50)  # Use blue circles for sensor nodes

    # Plot sinks as squares
    for sink_pos in sinks_positions:
        plt.plot(sink_pos[1], sink_pos[0], 's', color='red', markersize=10)  # Use red squares for sinks

    # Plot corrupted region as 'X'
    for corr_pos in corrupted_region:
        plt.text(corr_pos[1], corr_pos[0], 'X', color='black', fontsize=12, ha='center', va='center')

    # Plot the route between sensors
    plt.plot([route_start[1], route_end[1]], [route_start[0], route_end[0]], linestyle='-', color='green', linewidth=2, label='Original Route')

    # Plot the backup route avoiding the corrupted region
    plt.plot([backup_route[0][1], backup_route[1][1]], [backup_route[0][0], backup_route[1][0]], linestyle='--', color='orange', linewidth=2, label='Backup Route (Avoiding Corrupted Region)')

    # Set plot properties and legend
    plt.title('10x10 Grid with Sensors, Sinks, and Routes')
    plt.xticks(range(grid_size))
    plt.yticks(range(grid_size))
    plt.grid(True, linewidth=0.5, color='black', linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

# Define the grid size, sink positions, corrupted region, and routes
grid_size = 10
sinks_positions = [(0, 0), (0, grid_size-1), (grid_size-1, 0)]
corrupted_region = [(2, 2), (2, 3), (3, 2), (3, 3)]  # Example corrupted region
route_start = (5, 5)
route_end = (0, 0)
backup_route = [(5, 5), (9, 0)]

# Visualize the WSN with specified features
visualize_wsn(grid_size, sinks_positions, corrupted_region, route_start, route_end, backup_route)
```
